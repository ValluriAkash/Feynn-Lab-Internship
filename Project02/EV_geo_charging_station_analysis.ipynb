```python
# charging stations sanctioned visualization from dataset 1
plt.figure(figsize=(6, 6))
sns.barplot(data=df2, y=df2['State'].sort_values(ascending=True), x='No. of EV chargers sanctioned', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Number of Charging Stations', family='serif', fontsize=14, labelpad=10)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Number of Charging Stations Sanctioned in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

Maharashtra leads the way with 317 sanctioned EV chargers, followed by Andhra Pradesh (266) and Tamil Nadu (281). This distribution of sanctioned chargers indicates a focus on states with high EV adoption and population density. However, it is important to note that some states with a lower number of sanctioned chargers, such as Meghalaya and Sikkim, also demonstrate an intent to develop EV infrastructure. This dataset highlights the government's efforts to promote EV adoption by investing in charging infrastructure, which is crucial for enhancing the convenience and accessibility of EVs across the country.

```python
# National highway wise operational Public EV Charging Stations (PCS)
plt.figure(figsize=(16,4))
sns.barplot(data=df3, x='National Highway', y ='No. of operational PCS',palette='viridis')
sns.despine(right=False, top=False)
plt.tick_params(axis='x', rotation=90)
plt.xlabel('National Highway',family='serif', size=15)
plt.ylabel('Count', family='serif', size=15)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title('National highway wise operational Public EV Charging Stations (PCS)', family='serif', size=15)
plt.show()
```

Highways such as National Highway-44 have a significant number of operational PCS with 39 stations, followed by National Highway-16 with 37 stations and National Highway-65 with 22 stations. These highways likely represent major transportation routes and corridors with higher EV adoption and demand. On the other hand, some highways have a limited number of operational PCS, indicating the need for further infrastructure development to support EV usage in those regions. The presence of PCS along national highways is crucial for enabling long-distance travel and ensuring the accessibility and convenience of EV charging infrastructure for drivers.

```python
# retail outlets visualization from dataset - 3
plt.figure(figsize=(6, 8))
sns.pointplot(data=df4, x='Public EV Charging Stations (PCS)', y='EV Charging Facility', color='orange')
plt.xlabel('Number of Retail Outlets for Charging EVs', family='serif', size=12, labelpad=10)
plt.ylabel('State/Ut', family='serif', size=12)
plt.tick_params(direction='inout')
plt.xticks(family='serif', size=10)
plt.yticks(family='serif', size=10)
plt.title(label='Available Retail Outlets for Charging EVs in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```
